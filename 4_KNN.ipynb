# KNN

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

iris = datasets.load_iris()
```

```python
X_iris = iris.data[:, [2, 3]]
y_iris = iris.target
```

```python
plt.figure(figsize=(10, 7))

for cl, c in zip(np.unique(y_iris), ('green', 'blue', 'red')):
    plt.scatter(X_iris[y_iris==cl, 0], X_iris[y_iris==cl, 1], c=c)
plt.show()
```

```python
X = X_iris[y_iris != 0]
y = y_iris[y_iris != 0]


import matplotlib.pyplot as plt

plt.figure(figsize=(10 ,7))
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier


model_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', KNeighborsClassifier())
])

model_pipeline.fit(X_train, y_train)
```

```python
from plot_utils import MetricsUtils

metrics = MetricsUtils(x=X_train, y=y_train, estimator=model_pipeline)
metrics()
```

```python
metrics = MetricsUtils(x=X_test, y=y_test, estimator=model_pipeline)
metrics()
```
